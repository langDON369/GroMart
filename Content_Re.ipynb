```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=600)
```

```python
vector = cv.fit_transform(products['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[13]
```

```python
products[products['Name'] == 'Nutraj California Almonds (Badam)'].index[0]
```

```python
def recommend(item):
    index = products[products['Name'] == item].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(products.iloc[i[0]].Name)
        
```

```python
recommend('Sunny Sun Lite Refined Sunflower Oil')
```
